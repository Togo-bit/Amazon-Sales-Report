```python
null_columns = df.isnull().sum()
print(null_columns[null_columns > 0])
```

```python
df = df.dropna(subset = 'Amount')
df = df.dropna(subset = 'ship-city')
df = df.dropna(subset = 'ship-state')
df = df.dropna(subset = 'ship-postal-code')
df = df.dropna(subset = 'ship-country')
df
```

```python
df['promotion-ids'] = df['promotion-ids'].fillna('No ID found')
df['fulfilled-by'] = df['fulfilled-by'].fillna('No Record')
df
```

```python
df = df.dropna(subset = 'Unnamed: 22')
df.reset_index(inplace = True)
df
```

```python
df = df.drop('level_0',axis = 1)
df
```

```python
df[df['ship-state'] == 'ODISHA'].sort_values(by = 'Amount')
```

```python
df.groupby(['ship-state','ship-city'])['Amount'].agg(['sum','max','min','count','mean'])
```

```python
df.groupby('ship-state')['Amount'].agg(['sum','max','min','count','mean']).sort_values(by = 'sum',ascending = False)
```
